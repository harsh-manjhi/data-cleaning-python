```python
import pandas as pd
import numpy as np

data = {
    "customer_id":  [1, 2, 3, 4, 5, 6, 7, 8],
    "age":          [23, 45, "N/A", 31, np.nan, 52, 28, 33],
    "gender":       ["M", "Female", "male", "N/A", "F", "", "Male", "female"],
    "purchase_amt": [120.5, np.nan, 340.0, 89.0, 456.0, "null", 210.0, np.nan],
    "city":         ["Mumbai", "Delhi", "Mumbai", "?", "Chennai", "Delhi", "N/A", "Pune"],
    "subscribed":   ["Yes", "No", "Yes", "Yes", np.nan, "No", "Yes", "No"]
}

df = pd.DataFrame(data)
```

```python
df
```

##  Lets Check the Missing values detected by Pandas :

```python
df.isna().sum()
```

```python
df_clean = df.replace(["N/A",''," ", "null","?",], np.nan)
df_clean
```

```python
df_clean.isna().sum()
```

## - Missing data per Columns

```python
missing_data_pct = (df_clean.isna().sum() / df_clean.shape[0]) * 100

print("Missing data per Cols in %: ")
print(missing_data_pct)
```

## 1. 'Age' column in each column, we can easily replace the values with the median values as replacing the values with mean or average values will not be a good representative in a skewed you data.

```python
df_clean["age"]
```

## 2. In the "gender" column, as this column is a categorical feature, we can replace the values with mode or most frequent value - because that's the correct way of doing it

```python
df_clean["gender"]
```

## 3. Now the "purchased amount" column. This column is interesting because this would be the target variable. So for that reason I'll not be doing anything with this or simply will try to think to drop the rows with the null values.

```python
df_clean["purchase_amt"]
```

## 4. In the "city" column, as it is also a categorical attribute, I will replace the values with mode.

```python
df_clean["city"]
```

## 5. And the last "subscribed" column, I will replace the value with - I think mode cuz - I can see It's a binary column, but it's an also categorical column. So according to me, we should replace it with mode.

```python
df_clean["subscribed"]
```

```python
df_clean
```

## Starting with the first column: "Age"

```python
print(df_clean["age"].isna().sum())
df_clean["age"] = df_clean["age"].fillna(df_clean["age"].median())
df_clean["age"]
```

## it's time for "Gender" column

```python
df_clean["gender"].isna().sum()
```

```python
df_clean["gender"]
```

```python
df_clean["gender"].unique()
```

```python
df_clean["gender"] = df_clean["gender"].replace(['M','male'],"Male") # replacing male values
df_clean["gender"]
```

```python
df_clean["gender"] = df_clean["gender"] = df_clean["gender"].replace(['f',"female",'F'],"Female") # replacing female values
df_clean["gender"]
```

```python
df_clean["gender"] = df_clean["gender"].fillna(df_clean["gender"].mode()[0])
df_clean["gender"]
```

## Now "purchase amount" column

```python
df_clean
```

```python
df_clean_reduced = df_clean.dropna(subset=['purchase_amt'])
print("Purchase amt:\n", df_clean_reduced["purchase_amt"])
df_clean_reduced = df_clean_reduced.reset_index(drop = True)
df_clean_reduced
```

## cleaning the "city" attribute.

```python
df_clean_reduced["city"]
```

```python
df_clean_reduced["city"] = df_clean_reduced["city"].fillna(df_clean_reduced["city"].mode()[0])
df_clean_reduced
```

## now the "subscribed" column

```python
df_clean_reduced["subscribed"]
```

```python
df_clean_reduced["subscribed"].isna().sum()
```

```python
df_clean_reduced["subscribed_was_missing"] = df_clean_reduced["subscribed"].isna().astype(int)
df_clean_reduced["subscribed_was_missing"]
```

```python
df_clean_reduced["subscribed"] = df_clean_reduced["subscribed"].fillna(df_clean_reduced["subscribed"].mode()[0])
df_clean_reduced["subscribed"]
```

```python
# Verify - ing
df_clean_reduced["subscribed"].isna().sum()
```

```python
# Final verification
print("Final Missing values: ")
print()
print(df_clean_reduced.isna().sum())
print()

print("Shape: ",df_clean_reduced.shape)
print()
print("Dtypes: ",df_clean_reduced.dtypes)
```
